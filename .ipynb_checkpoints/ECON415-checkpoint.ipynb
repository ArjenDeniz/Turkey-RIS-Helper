# General Functions

```python
# to correct column names
def change_colnames(df,a):
    coln = df.loc[a,:].values.flatten().tolist()
    colNames = [str(col) for col in coln]
    df.columns = colNames
```

```python
# to sum years
def sum_years(df,start,year_gap = 10):
    a = df.iloc[:,start:(start+year_gap)].sum(axis=1)
    return a
```

```python
def totals(df,total_row,colnames):
    total_df = pd.DataFrame(total_row)
    total_df["Years"] = colnames
    total_df.drop(labels = [0,1],inplace = True)
    total_df.reset_index(inplace = True)
    total_df.drop(columns= "index", inplace = True)
    total_df.columns.values[[0]] = "Total"
    return total_df
```

```python
# region finder function
def region_finder(df, city_name, col_name, out_name, regional_cities, region_list):
    i = 0
    for reg in regional_cities:
        for city in city_name:
            if city in reg:
                    df.loc[df[col_name] == city,out_name] = region_list[i]
        i = i + 1
```

```python
# regional_sum function
def regional_sum(df,colReg_name,col_name,region_list):
    b = []
    for region in region_list:
        a = df.loc[df[colReg_name] == region, col_name].sum().to_list()
        b.append(a)
    
    total = pd.DataFrame(b,columns = col_name)
    total["NUTS-1 Region"] = region_list
    total.sort_values("NUTS-1 Region", inplace = True)
    return total
```

## Universities

```python
# data reading
uni = pd.read_csv("Üniversite Hakkında Genel Bilgiler.csv")
display(uni)
```

### Data Cleaning

```python
#fixing column names
change_colnames(uni,1)
```

```python
#cleaning empty rows & columns

# drop first 2 unuseful rows
uni.drop(labels=[0,1],inplace = True)

# drop empty rows
uni.dropna(subset = ["İl"],inplace = True)

#drop unnecessary columns
uni.reset_index(inplace = True)
uni.drop(columns = ["nan","index","Web Adresi","E-Posta","Telefon","Fax","Adres","Rektör"],axis = 1, inplace = True)
```

### University Age & Regions

```python
# finding the universities ages
current_year = 2022
uni['year'] = pd.DatetimeIndex(uni['Kuruluş Yılı']).year
uni['Age'] = [current_year- y for y in uni.year]
uni.drop(columns = ["Kuruluş Yılı","year"],inplace = True)
```

```python
#finding regions for universities
İl = uni.İl.to_list()
region_finder(uni, İl,"İl","NUTS-1 Region", regional_cities, region_list)
display(uni)
```

### Total Universities per Region

```python
# finding total universities for Regions

uni["Bölgesel Üniversite Sayısı"] = uni["NUTS-1 Region"].map(
    uni["NUTS-1 Region"].value_counts())
regional_uni = uni.drop_duplicates(subset="NUTS-1 Region")
regional_uni.sort_values("NUTS-1 Region", inplace=True)
regional_uni.reset_index(inplace=True)
regional_uni.drop(columns=["index", "İl", "Tür", "Birim Adı","Age","Bölge"], inplace=True)
display(regional_uni)
```
